```python
from lxml import etree
import pandas as pd

from tqdm import tqdm
import os

```

```python
#setup
path = 'OneDrive_1_14-03-2023/output_xml 2023-03-13 BR'
xml_file = 'DFT_20230313-164428-All Contracts.xml'
excel_file = '2023-03-11 BR MN Template for External Contracts - Service and Rental v1.9 - Brazil 2023-03-11.xlsx'
sheet_name = 'Service and Rental - TEMPLATE'
#load_file = '2023-03-06 BR Load Results.xlsx'
```

```python
df_log = pd.read_csv(f'{path}/logs.csv', header=None)
```

```python
# Opening excel and xml file
tree = etree.parse(f'{path}/{xml_file}')
root=tree.getroot()
excel = pd.read_excel(f'{path}/{excel_file}', header = 3, sheet_name=f'{sheet_name}')
#load = pd.read_excel(f'{path}/{load_file}')
```

```python
# saving removed products as list
df_log.columns=['t', 'd', 'a', 'c']
df_log[['A', 'B']] = df_log['a'].str.split(': ', 1, expand=True)
removed = df_log[['B']]
removed[['A', 'B']] = df_log['B'].str.split('_', 1, expand=True)
# for AZ
removed['A'] = removed['A'].str.replace("-", "")
removed = removed['A']
removed = list(removed.drop_duplicates(keep='last'))
```

```python
# removing products from log from excel
for index in tqdm(excel.index):
    if excel['Product ID'][index] in removed:
        excel.drop([index], axis=0, inplace=True)
```

```python
# count of products
excel_gr = excel.groupby(['Alternate Contract Id','Customer ID', 
'Price Program Name']).size().reset_index(name='counts')
```

```python
# Getting numbers and list with products

list_all = []
list_products =[]

Contract = tree.xpath('.//ContractOffer')
for contract in Contract:
    Program = contract.xpath('.//Program')
    for program in Program:
        nr_products = 0
        # if program.attrib['SourceStrategyID'] == 'AbbottTieredLineItemPricingStrategy':
        #     Product = program.xpath('.//ProductLI')
        #     for product in Product:
        #         list_all.append([contract.attrib['Customer'], contract.attrib['AlternateId'],program.attrib['Name'], product.attrib['ProductNum']])
        #         nr_products+=1
        # list_products.append([contract.attrib['Customer'], contract.attrib['AlternateId'],program.attrib['Name'], nr_products])
        
        if program.attrib['SourceStrategyID'] in ["AbbottServicePricingStrategy", "AbbottRentalPricingStrategy"]:
           Product = program.xpath('.//RentalLI')
           for product in Product:
                nr_products+=1
                list_all.append([contract.attrib['Customer'], contract.attrib['AlternateId'],program.attrib['Name'], product.attrib['ProductNum']])
        list_products.append([contract.attrib['Customer'], contract.attrib['AlternateId'],program.attrib['Name'], nr_products])
        
        
        # if program.attrib['SourceStrategyID']=="AbbottDOLPricingStrategy":
        #     Discount = program.xpath('.//DiscountOffListLI')
        #     nr_products = 0
        #     for discount in Discount:
        #         list_all.append([contract.attrib['Customer'], contract.attrib['AlternateId'],program.attrib['Name'], discount.attrib['ProductNum']])
        #         nr_products+=1
        #     list_products.append([contract.attrib['Customer'], contract.attrib['AlternateId'],program.attrib['Name'], nr_products])
        
            
```

```python
# Exporting to dataframe
df_xml = pd.DataFrame(list_products)
df_xml.columns =['Customer ID','AlternateId', 'Price Program Name', 'counts']
df_xml.sort_values('AlternateId')

df_all = pd.DataFrame(list_all)
df_all.columns =['Customer ID','AlternateId', 'Price Program Name', 'Product ID']




```

```python
df_xml
```

```python
# removing price programs and contract ids with -2
for line in tqdm(df_xml.index):
    if df_xml['Price Program Name'][line][-2:] == '-2':
        df_xml['Price Program Name'][line] = df_xml['Price Program Name'][line].replace("-2", "")
    if df_xml['AlternateId'][line][-2:] == '-2':
       df_xml['AlternateId'][line] = df_xml['AlternateId'][line].replace("-2", "")

df_xml = df_xml.groupby(['Customer ID','AlternateId', 'Price Program Name']).sum()
df_xml.reset_index(inplace=True)
```

```python
# number of all lines
print(excel.shape[0])
print(df_all.shape[0])
print(len(removed))
```

```python
# Merging dataframes- counts
excel_gr['Customer ID'] = excel_gr['Customer ID'].astype(int)
df_xml['Customer ID'] = df_xml['Customer ID'].astype(int)

df_merged = pd.merge(df_xml, excel_gr,  how='outer', left_on=['Customer ID','AlternateId', 'Price Program Name'], right_on = ['Customer ID','Alternate Contract Id', 'Price Program Name'])
```

```python
df_merged
df_merged.rename(columns = {'counts_x':'count from xml', 'counts_y':'count from excel'}, inplace = True)
```

```python
# checking if the same
df_merged['Are the same?'] = ''
for index in tqdm(df_merged.index):
    if df_merged['count from xml'][index] == df_merged['count from excel'][index]:
        df_merged['Are the same?'][index] = 'yes'
    else:
        df_merged['Are the same?'][index] = 'no'
```

```python
df_merged.to_excel(f'{path}/result.xlsx')
```
