```python
from lxml import etree
import pandas as pd
from tqdm import tqdm
import os

```

```python
#setup
path = 'OneDrive_2_19-04-2023'
xml_file= '20230421-100059-ALL CONTRACTS.xml'
# Opening excel and xml file
tree = etree.parse(f'{path}/{xml_file}')
root=tree.getroot()
```

```python
#####XML######
# Getting numbers and list with products
list_all = []
Contract = tree.xpath('.//ContractOffer')
for contract in Contract:
    Program = contract.xpath('.//Program')
    for program in Program:
        Discount = program.xpath('.//DiscountOffListLI')
        for discount in Discount:
            Price = discount.xpath('.//LIPrices')
            for price in Price:
                list_all.append([contract.attrib['Customer'], contract.attrib['DocName'],
                program.attrib['CustomerContractId'],
                # program.attrib['SalesDealNo'], 
                program.attrib['Name'],
                discount.attrib['ProductNum'],
                price.attrib['PriceHigh']])
# dataframes
xml_all = pd.DataFrame(list_all)
xml_all.columns =['Customer ID',  'Division','Contract ID',
'Price Program Name', 
'Product ID', 'Price']
```

```python
xml_all['Division2'] = xml_all['Division'].astype(str).str[:3]
```

```python
xml_crm = xml_all.loc[xml_all['Division']== 'CRM']
xml_nmd = xml_all.loc[xml_all['Division']== 'NMD']
```

```python
xml_rest = xml_all.loc[(xml_all['Division'] != 'NMD') & (xml_all['Division'] != 'CRM')]
```

```python
#####XML######
# Getting numbers and list with products
list_rebut = []
Contract = tree.xpath('.//ContractOffer')
for contract in Contract:
    Program = contract.xpath('.//Program')
    for program in Program:
        Line = program.xpath('.//LineItem')
        for line in Line:
            list_rebut.append([contract.attrib['CustomerContractId'], contract.attrib['DocName'],
            # program.attrib['SalesDealNo'], 
            program.attrib['Name'],
            line.attrib['ProductNum']])
# dataframes
xml_rebut = pd.DataFrame(list_rebut)
xml_rebut.columns =['Contract ID',  'Division',
'Price Program Name', 
'Product ID']
```
